````python
# time to read the sheet
df = pd.read_excel('.\\Raw Data\\Selected food price watch mar 2017- proshare.xlsx', sheet_name="NATIONAL")
# Time to get an idea about the raw data
df
````

# Needed steps
1. The actual column name is on the second row  
```1	ItemLabels	Unit of Measurement	2016-01-01 00:00:00	2016-02-01 00:00:00	...```

2. After fixing the column name, the first two rows are not part of the data.

3. The column heading read something like ```ItemLabels | Unit of Measurement | Jan 2016 | Feb 2016```

4. Only 2016 data are needed from the sheet, thus the last five columns need to be dropped

5. The last row of the data "%Average" is not needed


````python
#Fixing the column names and cleaning up of the index
rf_data = df.copy(); #getting copy of the real file
rf_data.columns = rf_data.iloc[1] #setting the colum name
rf_data.drop([0,1],inplace=True) #deleting the first two rows
rf_data.reset_index(inplace=True) #taking care of the soiled index due to the last operation
rf_data.drop(["index"], axis=1, inplace=True) # removing the previous index
rf_data.drop([rf_data.last_valid_index()],axis=0,inplace=True) # eliminating the last unwanted row

# Now, how does it look?
rf_data
#obviously we still need to tackle point 3 and 4 above

````

**Obviously we still need to tackle points 3 and 4 above, go to the next section**

````python
'''
For sure, the first two column are just strings.
Let's check the types of those colum names that looks like time.
I will pick one at random... say the fourth
'''
type(rf_data.columns[3])
````

**Wow it is of type datetime!!!**  
Since not all the column names are of the same type,
We have to make sure we attempt to convert only the ones of type 
*datetime.datetime* only,
Thus, we will need the datetime module.

Note: Although you may decide to import modules in the section where 
you first need it, 
but i is against the best practice, thus I have moved all imports to

````python
# creating a mapper for ensuring that only the datetime types get formated 
mapper = lambda x:x.strftime("%b %Y") if isinstance(x,datetime.datetime) else x
rf_data.columns = rf_data.columns.map(mapper)
#Time to get discard none 2016 columns i.e. the last five columns
price_2016=rf_data[rf_data.columns[0:-5]]

#How does it look?
price_2016
````

## Needed steps for the second data set
1. It looks identical to the previous raw data set, thus, 
all the previous codes will be reused with only one exception
2. Here, just the last four columns will be deleted not five as 
in the previous data set.

````python
#Fixing the column names and cleaning up of the index
rf_data = df2.copy(); #getting copy of the real file
rf_data.columns = rf_data.iloc[1] #setting the colum name
rf_data.drop([0,1],inplace=True) #deleting the first two rows
rf_data.reset_index(inplace=True) #taking care of the soiled index due to the last operation
rf_data.drop(["index"], axis=1, inplace=True) # removing the previous index
rf_data.drop([rf_data.last_valid_index()],axis=0,inplace=True) # eliminating the last unwanted row

#------------------

# creating a mapper for ensuring that only the datetime types get formated 
mapper = lambda x:x.strftime("%b %Y") if isinstance(x,datetime.datetime) else x
rf_data.columns = rf_data.columns.map(mapper)
#Time to get discard none 2016 columns i.e. the last five columns
price_2017_2019=rf_data[rf_data.columns[0:-4]]

#How does it look?
price_2017_2019
#-------------------
````

## Great,
next is to join the two data set to give a single sheet

The 2016 data consist of 25 rows(foods) while the 2017-2019 set consist of 43 foods
Here, we still need all the foods, thus we will end up with Nan values for 18 food prices 
throgh-out 2016

````python
price2016_2019 = price_2016.merge(price_2017_2019,how='outer')

#How does it look?
price2016_2019
````
